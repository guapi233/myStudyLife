# 字典 dict 

## 字典的定义

字典和列表的区别

-	列表是有序的
-	字典是无序的

字典用{}定义

字典使用  键值对  存储数据，键值对之间使用 ， 分隔

-	键key 是索引
-	值value 是数据
-	键和值之间使用：分隔
-	键必须是唯一的
-	值可以是是任何数据类型，但是键只能使用字符串，数字或者元组

## 字典的常见操作

xiao_dict = {}

### 取值 
dict[“key”] 

### 增加和修改


dict[“age”] = 18  # 不存在就新增

dict[“age”] =20  # 存在就是修改

### 删除

dict.pop(“key”)

### 统计，合并，清空

1.	len()
2.	update() 合并字典  #如果被合并的字典中包含已经存在的键值对，会覆盖原有的键值对
3.	clear()  清空字典

### 循环遍历


每次循环的都是Key

### 应用场景

使用多个键值对，存储描述一个物体的相关信息

将多个字典存放在一个列表中  （爬虫中也会使用）
